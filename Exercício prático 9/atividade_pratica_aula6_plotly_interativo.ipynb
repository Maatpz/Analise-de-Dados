## 4. Missão 1 — Barras: Receita por Canal

A missão prática da aula pede construir um gráfico de barras para responder:
**qual canal vende mais?**  
Os slides também sugerem usar o hover para injetar métricas secundárias sem poluir a tela. fileciteturn7file0

### Tarefa
1. Agregue a `receita` por `canal_venda`
2. Inclua também uma métrica secundária, como `quantidade`
3. Construa um gráfico de barras com Plotly Express
4. Ordene do maior para o menor
5. Faça o hover mostrar mais do que apenas a receita

### Perguntas
- Qual canal lidera a receita?
- O canal líder também lidera em quantidade?
- O hover ajudou a enriquecer a leitura sem poluir a tela?


```python
# Agregue receita e quantidade por canal_venda
df_canal = df.groupby('canal_venda').agg(
    receita=('receita', 'sum'),
    quantidade=('quantidade', 'sum')
).reset_index()

# Ordene por receita do maior para o menor
df_canal = df_canal.sort_values(by='receita', ascending=False)

# Construa um gráfico de barras com Plotly Express
fig = px.bar(
    df_canal,
    x='canal_venda',
    y='receita',
    title='Receita Total por Canal de Venda',
    labels={'canal_venda': 'Canal de Venda', 'receita': 'Receita Total'},
    hover_data={'quantidade': True, 'receita': ':.2f'}
)

# Mostra o gráfico
fig.show()
```

### Insight obrigatório

- O canal **Online** se destaca como o principal gerador de receita e volume de vendas, indicando sua forte performance e alcance.
- O canal de **Loja Física** apresenta a menor receita e quantidade, sugerindo que pode haver oportunidades para otimização ou que sua estratégia precise de revisão.
- Um gestor poderia investigar as táticas de marketing e vendas do canal Online para identificar melhores práticas e explorar se podem ser adaptadas para impulsionar os outros canais, especialmente a Loja Física. Além disso, uma análise de custo por canal poderia revelar a rentabilidade real de cada um.

## 5. Missão 2 — Linhas: Sazonalidade da Receita Mensal

Os slides destacam que o gráfico de linhas, com zoom e pan, é ideal para navegar no tempo e isolar picos. Também reforçam que o eixo temporal precisa estar corretamente tipado no Pandas. fileciteturn7file0

### Tarefa
1. Converta `data_venda` em data, se necessário
2. Agregue a `receita` por mês
3. Construa um gráfico de linha interativo
4. Teste o zoom nos meses de pico
5. Observe se novembro e dezembro se destacam

### Perguntas
- Existe sazonalidade?
- Quais meses chamam atenção?
- O zoom ajudou a explorar melhor a parte final da série?


```python
# Agregando a receita por mês de forma robusta
# Criamos uma representação de período (Ano-Mês) para evitar erros de índice
df_mensal = df.groupby(df['data_venda'].dt.to_period('M'))['receita'].sum().reset_index()

# Convertendo de volta para timestamp para que o Plotly reconheça como eixo temporal
df_mensal['data_venda'] = df_mensal['data_venda'].dt.to_timestamp()

# Renomear para clareza
df_mensal = df_mensal.rename(columns={'data_venda': 'mes'})

# Construa o gráfico de linha interativo
fig = px.line(
    df_mensal,
    x='mes',
    y='receita',
    title='Sazonalidade da Receita Mensal',
    labels={'mes': 'Mês', 'receita': 'Receita Total'},
    markers=True
)

# Ajustes de layout para interatividade
fig.update_layout(
    xaxis_title='Mês',
    yaxis_title='Receita Total',
    hovermode='x unified'
)

fig.show()
```

### Insight obrigatório
Explique:
- qual tendência aparece
- se há picos claros
- qual hipótese de negócio pode explicar o comportamento observado


## 6. Missão 3 — Dispersão: Lucro vs. Receita segmentado por Categoria

Os slides mostram o uso do gráfico de dispersão para revelar correlação entre KPIs e identificar anomalias, com grande apoio do hover. fileciteturn7file0

### Tarefa
1. Construa um scatter plot com:
   - eixo X = `receita`
   - eixo Y = `lucro`
2. Use `categoria` como cor
3. Inclua no hover:
   - `produto`
   - `canal_venda`
   - `uf`
   - `margem_lucro`
4. Tente identificar algum ponto fora da curva

### Perguntas
- Existe correlação visual entre receita e lucro?
- Há anomalias?
- O hover ajuda a transformar um ponto em um caso investigável?


```python
# Construa o scatter plot interativo entre receita e lucro
fig = px.scatter(
    df,
    x='receita',
    y='lucro',
    color='categoria',
    title='Análise de Lucro vs. Receita por Categoria',
    labels={
        'receita': 'Receita (R$)',
        'lucro': 'Lucro (R$)',
        'categoria': 'Categoria',
        'margem_lucro': 'Margem de Lucro',
        'canal_venda': 'Canal de Venda',
        'uf': 'UF'
    },
    hover_data=['produto', 'canal_venda', 'uf', 'margem_lucro'],
    template='plotly_white'
)

# Melhora o layout e a legibilidade
fig.update_traces(marker=dict(size=8, opacity=0.7))
fig.update_layout(hovermode='closest')

# Exibe o gráfico
fig.show()
```

### Insight obrigatório
- **Correlação:** Observa-se uma correlação positiva clara entre receita e lucro, onde o aumento das vendas geralmente acompanha um aumento no lucro.
- **Anomalias:** Existem pontos (como em Informática ou Móveis) que fogem da linha principal, indicando produtos com margens muito acima ou abaixo da média da categoria.
- **Ação Analítica:** O hover permitiu identificar produtos específicos e UFs com baixa rentabilidade. A próxima ação seria investigar se esses casos de baixo lucro com alta receita são fruto de descontos excessivos ou custos logísticos elevados.

## 7. Exploração espacial — Onde está concentrada a operação?

Os slides incluem mapas geográficos como resposta para perguntas espaciais como:
**onde está concentrada nossa operação?** fileciteturn7file0

### Tarefa
Usando `latitude` e `longitude`, crie um mapa interativo que ajude a visualizar a operação.

### Sugestões
- use tamanho ou cor para uma métrica relevante (como receita)
- teste o zoom do mapa
- observe se há concentração regional

### Perguntas
- Onde a operação parece mais concentrada?
- O mapa ajudou mais do que uma simples tabela por UF?


```python
# Criando o mapa de dispersão geográfica
fig = px.scatter_geo(
    df,
    lat='latitude',
    lon='longitude',
    color='categoria',
    size='receita',
    hover_name='produto',
    hover_data=['uf', 'receita', 'canal_venda'],
    title='Concentração Geográfica das Vendas (por Receita)',
    scope='south america',
    template='plotly_white'
)

# Ajustando o foco do mapa para o Brasil
fig.update_geos(
    lataxis_range=[-35, 5],
    lonaxis_range=[-75, -35]
)

fig.show()
```

### Análise Geográfica
- **Concentração:** A operação mostra uma presença nacional, mas com pontos de maior volume (círculos maiores) concentrados no litoral e em capitais, facilitando a visualização da logística.
- **Vantagem Visual:** Diferente de uma tabela, o mapa permite identificar a proximidade entre vendas de diferentes estados e o alcance real da marca no território brasileiro.
- **Interatividade:** O zoom permite investigar cidades específicas ou regiões onde a densidade de pedidos é menor, ajudando na tomada de decisão sobre expansão regional.

## 9. Plotly Express — Máximo impacto, mínimo código

A aula mostra o Plotly Express como uma API declarativa, perfeita para DataFrames do Pandas e com geração automática de HTML/JS. fileciteturn7file0

### Tarefa
Explique em markdown:
1. O que significa dizer que Plotly Express é “declarativo”?
2. Em que ele facilita a vida do analista?
3. Como isso se conecta com a ideia de futuro dashboard em Streamlit?


### Respostas da Tarefa

1. **O que significa dizer que Plotly Express é “declarativo”?**
   Significa que o analista descreve *o que* quer ver (ex: 'quero um gráfico de barras com X e Y') em vez de programar *como* o computador deve desenhar cada linha, pixel ou eixo. Você declara as variáveis do DataFrame que alimentam as propriedades do gráfico, e a biblioteca cuida de toda a renderização complexa de HTML e JavaScript nos bastidores.

2. **Em que ele facilita a vida do analista?**
   Ele reduz drasticamente a quantidade de código (boilerplate). Com uma única linha, é possível criar gráficos complexos com legendas automáticas, escalas de cores, facetamento (subplots) e, principalmente, interatividade nativa (hover, zoom). Isso permite que o foco seja na análise dos dados e não na sintaxe da biblioteca de visualização.

3. **Como isso se conecta com a ideia de futuro dashboard em Streamlit?**
   Gráficos do Plotly são objetos Python que podem ser renderizados diretamente em aplicações web como o Streamlit através de comandos simples (como `st.plotly_chart(fig)`). Como ambos seguem a filosofia de 'Python puro' para gerar interfaces ricas, o que você desenvolve hoje no notebook pode ser copiado e colado em um script Streamlit para virar um dashboard profissional em minutos.
