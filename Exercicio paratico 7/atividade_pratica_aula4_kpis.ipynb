## 4. KPIs globais da operação

Com base nas “4 Métricas de Ouro do Varejo” apresentadas na aula, calcule os seguintes KPIs globais da base inteira: fileciteturn5file0

- **Receita Total**
- **Lucro Total**
- **Margem de Lucro (%)**
- **Ticket Médio**

### Orientação
- Pense primeiro na fórmula de cada KPI
- Escreva os cálculos em células separadas ou em uma pequena tabela-resumo

```python
# Calcule aqui os 4 KPIs globais

# 1. Receita Total
receita_total = df['receita'].sum()

# 2. Lucro Total
lucro_total = df['lucro'].sum()

# 3. Margem de Lucro (%)
margem_lucro_percentual = (lucro_total / receita_total) * 100

# 4. Ticket Médio
total_pedidos = df['pedido_id'].nunique() # Número de pedidos únicos
ticket_medio = receita_total / total_pedidos

# Exibir os resultados formatados
print(f"KPIs Globais da Operação:\n")
print(f"Receita Total: R$ {receita_total:,.2f}")
print(f"Lucro Total:   R$ {lucro_total:,.2f}")
print(f"Margem de Lucro: {margem_lucro_percentual:.2f}%")
print(f"Ticket Médio:  R$ {ticket_medio:,.2f}")
```

## 7. Análise por canal

### Pergunta de negócio
**Qual canal de venda é o mais eficiente?**

### Tarefa
Crie uma tabela agregada por `canal_venda` contendo, no mínimo:
- Receita Total
- Lucro Total
- Margem Média ou Margem Calculada
- Quantidade de pedidos
- Ticket Médio

### Depois responda
1. Qual canal gera mais receita?

2. Qual canal gera maior eficiência?

3. Se você fosse gestor, priorizaria volume ou margem?




```python
# Gere aqui a tabela agregada por canal_venda

análise_por_canal = df.groupby('canal_venda').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('pedido_id', 'nunique')
).reset_index()

análise_por_canal['margem_lucro_percentual'] = (análise_por_canal['lucro_total'] / análise_por_canal['receita_total']) * 100
análise_por_canal['ticket_medio'] = análise_por_canal['receita_total'] / análise_por_canal['quantidade_pedidos']

display(análise_por_canal.sort_values(by='receita_total', ascending=False))
```

## 9. Análise por região (UF)

### Pergunta de negócio
**Quais UFs concentram a operação?**

### Tarefa
Monte uma tabela por `uf` com:
- Receita Total
- Lucro Total
- Participação percentual da receita
- Ticket Médio

### Depois responda
1. Quais UFs concentram maior receita?

```python
# Gere aqui a análise por UF
análise_por_uf = df.groupby('uf').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('pedido_id', 'nunique')
).reset_index()

# Calcular a participação percentual da receita de cada UF
receita_total_geral = análise_por_uf['receita_total'].sum()
análise_por_uf['participacao_receita_percentual'] = (análise_por_uf['receita_total'] / receita_total_geral) * 100

# Calcular Ticket Médio
análise_por_uf['ticket_medio'] = análise_por_uf['receita_total'] / análise_por_uf['quantidade_pedidos']

# Calcular Margem de Lucro Percentual
análise_por_uf['margem_lucro_percentual'] = (análise_por_uf['lucro_total'] / análise_por_uf['receita_total']) * 100

display(análise_por_uf.sort_values(by='receita_total', ascending=False))
```

2.  **Existe concentração regional?**
    Sim, existe uma concentração regional notável. Observamos que as UFs do Sudeste (SP, RJ, MG) e Nordeste (PE) lideram em receita, indicando que a maior parte da operação está concentrada nessas regiões. As demais UFs têm participações menores e mais distribuídas.

3.  **Se fosse necessário expandir ou reforçar logística, por onde você começaria?**
    Com base na concentração de receita, a expansão ou reforço logístico deveria começar pelas UFs com maior participação, como **São Paulo (SP)** e **Rio de Janeiro (RJ)**, para otimizar o atendimento onde o volume é maior. Além disso, seria prudente analisar **Pernambuco (PE)** e **Minas Gerais (MG)** para garantir que a infraestrutura nessas regiões suporte o volume atual e potencial crescimento, considerando sua relevância em receita.

## 10. O motor da análise: agregações (groupby)

Nos slides, a lógica central é:
- `receita -> sum`
- `margem -> mean` ou cálculo derivado
- `produto -> count`
fileciteturn5file0

### Tarefa
Escolha **duas dimensões diferentes** e construa tabelas agregadas mostrando que você entendeu essa lógica.

Exemplos:
- por canal
- por categoria
- por UF
- por mês


```python
# Demonstre aqui o uso de groupby com diferentes agregações

# Garante que a coluna 'data_venda' está no formato datetime
df['data_venda'] = pd.to_datetime(df['data_venda'])

# Agregação por canal_venda
print("\n--- Análise por Canal de Venda ---")
análise_canal_agg = df.groupby('canal_venda').agg(
    receita_total=('receita', 'sum'),
    margem_lucro_media=('margem_lucro', 'mean'),
    total_pedidos=('pedido_id', 'nunique')
).reset_index()
display(análise_canal_agg.sort_values(by='receita_total', ascending=False))

# Agregação por mês
print("\n--- Análise Mensal ---")
df['mes'] = df['data_venda'].dt.month # Extrai o mês da data
análise_mes_agg = df.groupby('mes').agg(
    receita_total=('receita', 'sum'),
    margem_lucro_media=('margem_lucro', 'mean'),
    total_pedidos=('pedido_id', 'nunique')
).reset_index()
display(análise_mes_agg.sort_values(by='receita_total', ascending=False))
```

## 11. Campeões e vilões (Top N)

A aula destaca o poder dos rankings curtos e focados. fileciteturn5file0

### Tarefa
Crie:
- Top 10 produtos por **lucro**

### Perguntas
1. Quem são os “campeões”?

```python
# Monte aqui os rankings Top N

# Top 10 produtos por lucro
top_10_produtos_lucro = df.groupby('produto').agg(
    lucro_total=('lucro', 'sum')
).reset_index().sort_values(by='lucro_total', ascending=False).head(10)
print("\n--- Top 10 Produtos por Lucro ---")
display(top_10_produtos_lucro)

# Top 10 produtos por receita
top_10_produtos_receita = df.groupby('produto').agg(
    receita_total=('receita', 'sum')
).reset_index().sort_values(by='receita_total', ascending=False).head(10)
print("\n--- Top 10 Produtos por Receita ---")
display(top_10_produtos_receita)

# 5 Categorias com pior margem
piores_5_categorias_margem = df.groupby('categoria').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum')
).reset_index()
piores_5_categorias_margem['margem_lucro_percentual'] = (piores_5_categorias_margem['lucro_total'] / piores_5_categorias_margem['receita_total']) * 100
piores_5_categorias_margem = piores_5_categorias_margem.sort_values(by='margem_lucro_percentual', ascending=True).head(5)
print("\n--- 5 Categorias com Pior Margem de Lucro ---")
display(piores_5_categorias_margem)

# 5 UFs com pior margem
piores_5_uf_margem = df.groupby('uf').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum')
).reset_index()
piores_5_uf_margem['margem_lucro_percentual'] = (piores_5_uf_margem['lucro_total'] / piores_5_uf_margem['receita_total']) * 100
piores_5_uf_margem = piores_5_uf_margem.sort_values(by='margem_lucro_percentual', ascending=True).head(5)
print("\n--- 5 UFs com Pior Margem de Lucro ---")
display(piores_5_uf_margem)
```

## 12. Análise temporal

### Tarefa

   - mês

   - receita

### Depois responda

```python
# Faça aqui a análise temporal mensal

# 1. Converta data_venda para data, se necessário
df['data_venda'] = pd.to_datetime(df['data_venda'])

# 2. Crie colunas auxiliares
df['mes'] = df['data_venda'].dt.month
df['trimestre'] = df['data_venda'].dt.quarter

# 3. Gere uma tabela agregada por mês
análise_temporal_mensal = df.groupby('mes').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
    quantidade_pedidos=('pedido_id', 'nunique') # Para calcular o ticket médio
).reset_index()

análise_temporal_mensal['ticket_medio'] = análise_temporal_mensal['receita_total'] / análise_temporal_mensal['quantidade_pedidos']
análise_temporal_mensal['margem_lucro_percentual'] = (análise_temporal_mensal['lucro_total'] / análise_temporal_mensal['receita_total']) * 100

display(análise_temporal_mensal.sort_values(by='mes'))
```

Transformar linhas de dados limpos em **apoio real para decisões de negócio**. fileciteturn5file0


```python
# Desafio extra opcional

# 1. Tabela executiva final com KPIs globais
global_kpis = pd.DataFrame({
    'KPI': ['Receita Total', 'Lucro Total', 'Margem de Lucro (%)', 'Ticket Médio'],
    'Valor': [
        f"R$ {df['receita'].sum():,.2f}",
        f"R$ {df['lucro'].sum():,.2f}",
        f"{(df['lucro'].sum() / df['receita'].sum()) * 100:.2f}%",
        f"R$ {(df['receita'].sum() / df['pedido_id'].nunique()):,.2f}"
    ]
})
print("--- KPIs Globais ---")
display(global_kpis)

# 2. Tabela por canal com os mesmos indicadores
kpis_por_canal = df.groupby('canal_venda').agg(
    receita_total=('receita', 'sum'),
    lucro_total=('lucro', 'sum'),
    total_pedidos=('pedido_id', 'nunique')
).reset_index()

kpis_por_canal['margem_lucro_percentual'] = (kpis_por_canal['lucro_total'] / kpis_por_canal['receita_total']) * 100
kpis_por_canal['ticket_medio'] = kpis_por_canal['receita_total'] / kpis_por_canal['total_pedidos']

# Formatar para exibição
kpis_por_canal['receita_total'] = kpis_por_canal['receita_total'].apply(lambda x: f"R$ {x:,.2f}")
kpis_por_canal['lucro_total'] = kpis_por_canal['lucro_total'].apply(lambda x: f"R$ {x:,.2f}")
kpis_por_canal['margem_lucro_percentual'] = kpis_por_canal['margem_lucro_percentual'].apply(lambda x: f"{x:.2f}%")
kpis_por_canal['ticket_medio'] = kpis_por_canal['ticket_medio'].apply(lambda x: f"R$ {x:,.2f}")

print("\n--- KPIs por Canal de Venda ---")
display(kpis_por_canal.set_index('canal_venda'))
```

### Comparação e Interpretação

**O KPI global esconde alguma diferença importante entre os canais?**

Sim, o KPI global de margem de lucro (28.34%) esconde diferenças importantes entre os canais. Embora a margem global seja saudável, ao analisar por canal, percebemos que o **canal Online** possui a maior margem (29.56%), indicando ser o mais eficiente. Em contraste, a **Loja Física** apresenta a menor margem (27.24%).

Essa diferença, mesmo que pareça pequena em termos percentuais, representa uma variação significativa na rentabilidade por venda e na eficiência operacional de cada canal. Um gestor que olhasse apenas para a margem global poderia não identificar a oportunidade de otimizar os canais de menor margem ou de replicar as melhores práticas do canal mais eficiente.
